```python
BASE_URL = 'https://www.poynter.org/ifcn-covid-19-misinformation/page/'
```

```python
options = ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--disable-gpu')

wd = webdriver.Chrome(options=options)
```

```python
data = []
for i in range(1,765):
  url = f'{BASE_URL}{str(i)}'
  print(url)
  wd.get(url)
  pause_time =  randint(10,40)
  wait = WebDriverWait(wd,pause_time)
  data_scrapped = wait.until(EC.presence_of_all_elements_located((By.XPATH,"/html/body/div[1]/div[4]/div[2]/main/div/div[3]/div")))
  date_country = data_scrapped[0].find_elements_by_tag_name('strong')
  class_tweets = data_scrapped[0].find_elements_by_tag_name('h2')
  for i in range(len(date_country)):
    arr=[]
    inf = date_country[i].text.split('| ')
    print(inf)
    arr.append(inf[0])
    try:
      arr.append(inf[1])
    except:
      print('error aquii')
      arr.append('')
    inf2 = class_tweets[i].text.split(': ')
    arr.append(inf2[0].lower())
    arr.append(inf2[1])
    data.append(arr)
    print(inf2)
  clear_output()
  sleep(10)
```

```python
def prepare_tweet(tweet):
    try:        
        indices = [i for i, x in enumerate(tweet) if x == "'"]
        if indices == []:
            return tweet
        var = 0
        for a in indices:
            tweet = tweet[:a+var] + "'" + tweet[a+var:]
            var = var + 1
        return tweet
    except:
        return tweet
```

```python
def call_sp_ins_tweets_ifcn(data):
    conn = get_connection_covid_coaid()
    line_count = 0
    for element in data:
        cur = conn.cursor()
        try:
            call_sp = f"CALL public.insert_infc_tweet('{prepare_tweet(element[3])}', '{element[0]}', '{prepare_tweet(element[2])}', '{prepare_tweet(element[1])}')"
            cur.execute(call_sp)
            conn.commit()
            line_count=line_count+1
            if line_count%1000 == 0:
                print(f'Processed {line_count} lines.')
        except(Exception, psycopg2.DatabaseError) as error:
            print(error)
        finally:
            cur.close()
```

```python
call_sp_ins_tweets_ifcn(data)
```

```python
def get_data_class_view():
    conn = get_connection_covid_coaid()
    try:
        cur = conn.cursor()
        cur.execute('select * from get_data_class_view ')
        colnames = [desc[0] for desc in cur.description]
        data = cur.fetchall()
        return data,colnames
    except(Exception, psycopg2.DatabaseError) as error:
        print(error)
data,colnames = get_data_class_view()
```

```python
df = pd.DataFrame(data,
                  columns= colnames)
df.index = pd.to_datetime(df['tweet_created_at'])
df = df.groupby([df.index.year.values,
                 df.index.month.values,
                df['tweet_class'].values]).count()
df
```
